```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/city wise op.csv")
df
```

```python
df
```

```python
df['Total'] = df.iloc[:, 1:].sum(axis=1)
df

```

```python
# prompt: Using dataframe df: bar plot district wise

# Assuming 'DISTRICT ' column contains district names
# And you want to plot a bar chart for 'Total'
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(df['DISTRICT '], df['Total'])
plt.xlabel('District')
plt.ylabel('Total')
plt.title('District-wise Total')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
